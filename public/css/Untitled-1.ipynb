In this video, we’ll provide the Project Scenario and Overview. The commercial space age is here, with companies making space travel affordable for everyone. Virgin Galactic offers suborbital spaceflights, Rocket Lab provides small satellite services, and Blue Origin manufactures reusable rockets. Among these, SpaceX stands out as the most successful.
Some of SpaceX’s accomplishments include sending spacecraft to the International Space Station, launching Starlink, a satellite internet constellation, and sending manned missions to space. A key factor in SpaceX’s success is the relatively inexpensive cost of their rocket launches. For example, SpaceX advertises Falcon 9 rocket launches on its website at a cost of $62 million, while other providers cost upwards of $165 million each. This significant savings is largely due to SpaceX’s ability to reuse the first stage of the rocket.
To better understand the scale of the Falcon 9, we’ll use diagrams from Forest Katsch, a 3D artist and software engineer who creates infographics on spaceflight and spacecraft art. Katsch also develops software. The payload is enclosed in the fairings, and the second stage, or stage two, helps bring the payload to orbit. However, most of the work is done by the first stage, which is shown here. This stage is much larger than the second stage and performs most of the rocket’s tasks.
In this capstone project, you’ll take on the role of determining **the cost of a Falcon 9 launch** based on whether the first stage will land. Sometimes, the first stage doesn’t land and crashes, as shown in this clip. Other times, SpaceX may sacrifice the first stage due to mission parameters such as payload, orbit, and customer requirements.
As a data scientist working for Space Y, a new rocket company founded by billionaire industrialist Allon Musk, you’ll be tasked with determining the price of each launch. Your mission is to gather information about SpaceX and create dashboards for your team. Additionally, you’ll determine whether SpaceX will reuse its first stage. Instead of relying solely on rocket science to predict the successful landing of the first stage, you’ll train a machine learning model using public information to make this prediction.
